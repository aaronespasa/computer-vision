```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Read in the image
image = cv2.imread("Images/city-hall.jpg")

# Make a copy of the image
image_copy = np.copy(image)

# Chage color to RGB (from BGR)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
# Convert to grayscale for filtering
gray_image = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

plt.imshow(gray_image, cmap='gray')
```

```python
# 3x3 array (kernel) for edge detection (To detect vertical edges and ignore horizontal edges)
# Sobel filter
sobel_x = np.array([[ -1, 0, 1],
                   [ -2, 0, 2],
                   [ -1, 0, 1]])

# Perform convolution using filter2D, which has inputs: (grayscale image, bit-depth, kernel)
filtered_image = cv2.filter2D(gray_image, -1, sobel_x)

plt.imshow(filtered_image, cmap="gray")
```

```python
# To create the binary image:
# Create threshold that sets all the filtered pixels to white
# Above a certain threshold
# cv2.threshold(filtered_image, low_value_for_the_pixel_values_to_pass_through, pixel_value_to_tur_these_all_into, type_threshold)

retval, binary_image = cv2.threshold(filtered_image, 150, 255, cv2.THRESH_BINARY)

plt.imshow(binary_image, cmap="gray")
```
